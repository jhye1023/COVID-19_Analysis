```python
import pandas as pd
data = pd.read_json(path_or_buf ='https://api.covidtracking.com/v1/us/daily.json')
data.info()
```

```python
data.head()
```

```python
data.plot()
```

```python
data.columns
```

```python
increase_data = data.loc[:,['date','deathIncrease','hospitalizedIncrease', 'negativeIncrease', 'positiveIncrease']]
increase_data.head()
```

```python
final_data = increase_data.astype({'date':'str'})
final_data.info()
```

```python
final_data.set_index('date', inplace=True)
final_data.head()
```

```python
final_data.plot()
```

```python
final_data.sort_values(by='date', inplace=True)
final_data.plot()
```

```python
final_data.plot(subplots=True)
```

```python
final_data[-15:].plot()
```

```python
final_data[-15:].plot.bar(stacked=True)
```

```python
final_data.plot.bar(stacked=True)
```

```python
final_data.plot(subplots=True, layout=(2,2))
```
